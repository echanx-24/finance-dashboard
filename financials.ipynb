```python
connection = sql.connect("dashboard/financials.db")
cursor = connection.cursor()
```

```python
df = pd.read_sql("SELECT * FROM Income_Statement_Qtr WHERE ticker == 'BB'", connection)
```

```python
# elements = [str(i) + " " + str(j) for i, j in zip(df["YEAR"], df["FRAME"])]
# df["TIMEFRAME"] = elements
df = df[["TIMEFRAME", "REVENUE", "COSTOFREVENUE", "GROSSPROFIT",
   "RESEARCH", "SGA", "OTHEROPERATING", "OPERATINGEXP", "OPERATINGINC"]]
df = df.set_index("TIMEFRAME")
```

```python
df2 = df.iloc[::-1].T
df2
```

```python
for i in df2.columns:
    if df2[i].dtype == "float64" or df2[i].dtype == "int64":
        df2[i] = df2[i]/1000
        df2[i] = [f"{int(j):,d}" for j in df2[i]]
```
